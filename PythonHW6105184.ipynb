```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Read file

```python
file=pd.read_csv('COVID19.csv')
file
```

```python
#Infected
ThI=0 #Thailand
ArI=0 #Argentina
SeI=0 #Sweden
EsI=0 #Spain
KrI=0 #South Korea
SgI=0 #Singapore
PtI=0 #Portugal
NlI=0 #Netherlands
JpI=0 #Japan
ItI=0 #Italy
```

```python
#Death
ThD=0 #Thailand
ArD=0 #Argentina
SeD=0 #Sweden
EsD=0 #Spain
KrD=0 #South Korea
SgD=0 #Singapore
PtD=0 #Portugal
NlD=0 #Netherlands
JpD=0 #Japan
ItD=0 #Italy
```

```python
#Collected or Dum aray
ThC=[] #Thailand
ArC=[] #Argentina
SeC=[] #Sweden
EsC=[] #Spain
KrC=[] #South Korea
SgC=[] #Singapore
PtC=[] #Portugal
NlC=[] #Netherlands
JpC=[] #Japan
ItC=[] #Italy
```

```python
for b in range(10332):
    a=10331-b
    country=file.iloc[a]['countriesAndTerritories']
    deaths=file.iloc[a]['deaths']
    cases=file.iloc[a]['cases']
    if country == 'Thailand' :
        ThD=ThD+deaths
        ThI=ThI+cases
        ThC.append(ThI)
    if country == 'Argentina' :
        ArD=ArD+deaths
        ArI=ArI+cases
        ArC.append(ArI)
    if country == 'Sweden' :
        SeD=SeD+deaths
        SeI=SeI+cases
        SeC.append(SeI)
    if country == 'Spain' :
        EsD=EsD+deaths
        EsI=EsI+cases
        EsC.append(EsI)
    if country == 'South_Korea' :
        KrD=KrD+deaths
        KrI=KrI+cases
        KrC.append(KrI)
    if country == 'Singapore' :
        SgD=SgD+deaths
        SgI=SgI+cases
        SgC.append(SgI)
    if country == 'Portugal' :
        PtD=PtD+deaths
        PtI=PtI+cases
        PtC.append(PtI)
    if country == 'Netherlands' :
        NlD=NlD+deaths
        NlI=NlI+cases
        NlC.append(NlI)
    if country == 'Japan' :
        JpD=JpD+deaths
        JpI=JpI+cases
        JpC.append(JpI)
    if country == 'Italy' :
        ItD=ItD+deaths
        ItI=ItI+cases
        ItC.append(ItI)
```

```python
#Number of case
NumCase=[ThI,ArI,SeI,EsI,KrI,SgI,PtI,NlI,JpI,ItI]
NumCase
```

```python
#Percent of death of case
PDTh=ThD*100/ThI
PDAr=ArD*100/ArI
PDSe=SeD*100/SeI
PDEs=EsD*100/EsI
PDKr=KrD*100/KrI
PDSg=SgD*100/SgI
PDPt=PtD*100/PtI
PDNl=NlD*100/NlI
PDJp=JpD*100/JpI
PDIt=ItD*100/ItI
PerDeath=[PDTh,PDAr,PDSe,PDEs,PDKr,PDSg,PDPt,PDNl,PDJp,PDIt]
PerDeath
```

```python
#Percent of case
PCTh=ThI*100/69428524
PCAr=ArI*100/44494502
PCSe=SeI*100/10183175
PCEs=EsI*100/46723749
PCKr=KrI*100/51635256
PCSg=SgI*100/5638676
PCPt=PtI*100/10281762
PCNl=NlI*100/17231017
PCJp=JpI*100/126529100
PCIt=ItI*100/60431283
PerCase=[PCTh,PCAr,PCSe,PCEs,PCKr,PCSg,PCPt,PCNl,PCJp,PCIt]
PerCase
```

```python
fig=plt.figure()
plt.plot(np.arange(len(ThC))+1,ThC,label='Thailand')
plt.plot(np.arange(len(ArC))+1,ArC,label='Argentina')
plt.plot(np.arange(len(SeC))+1,SeC,label='Sweden')
plt.plot(np.arange(len(EsC))+1,EsC,label='Spain')
plt.plot(np.arange(len(KrC))+1,KrC,label='South Korea')
plt.plot(np.arange(len(SgC))+1,SgC,label='Sigapore')
plt.plot(np.arange(len(PtC))+1,PtC,label='Portugal')
plt.plot(np.arange(len(NlC))+1,NlC,label='Netherlands')
plt.plot(np.arange(len(JpC))+1,JpC,label='Japan')
plt.plot(np.arange(len(ItC))+1,ItC,label='Italy')

plt.yscale('log')
plt.legend(bbox_to_anchor=(1.1,0.9),loc='upper left')
plt.title("Relation graph between days and log of cases number in 10 countries")
plt.xlabel("Days")
plt.ylabel("Log of cases number")
plt.xlim(0)
```
