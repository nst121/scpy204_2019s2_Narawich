```python
import numpy as np
import matplotlib.pyplot as plt
from astroquery.sdss import SDSS
```

```python
query="SELECT s.z, s.zErr, p.petroMag_g, p.petroMagErr_g    FROM PhotoObjAll p   JOIN SpecObjAll s ON s.bestObjID=p.objID   WHERE s.zWarning=0 and s.z between 0.3 and 0.4 and p.petroMag_g>-5000 and p.petroMagErr_g<1 and s.zErr<=0.0002" 
m=SDSS.query_sql(query)
print(m)
```

```python
z=m["z"]
zErr=m["zErr"]
petroMag_g=m["petroMag_g"]
petroMagErr_g=m["petroMagErr_g"]
```

```python
fig=plt.figure()
plt.errorbar(z, petroMag_g, yerr=petroMagErr_g, xerr=zErr , fmt='o', markersize=2.0)
plt.xlim(0.3,0.4)
plt.title("The graph of relation between petroMag_g and z")
plt.xlabel("z")
plt.ylabel("petroMag_g")
```
